```python
def GetTitle(name):
    title_group = {
        'mr' : 'Mr',
         'mrs' : 'Mrs',
        'miss': 'Miss',
        'master': 'Master',
        'don' : 'Sir',
        'rev' : 'Sir',
        'dr'  : 'officer',
        'mme' : 'Mrs',
        'ms'  : 'Mrs',
        'major' : 'Officer',
        'lady' : 'Lady',
        'sir' :'Sir',
        'mlle':'Miss',
        'col' : 'Officer',
        'capt' : 'Officer',
        'the countness' : 'Lady',
        'jonkheer' : 'Sir',
        'dona':'Lady',
        'the countess': 'Lady'
    }
    
    first_name_with_title = name.split(',')[1]
    title = first_name_with_title.split('.')[0]
    title = title.strip().lower()
    return title_group[title]
```

```python
df.Name.map(lambda x : GetTitle(x)).unique()
```

```python
df['Title'] = df.Name.map(lambda x: GetTitle(x))
```

```python
df.head()
```
